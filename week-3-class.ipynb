The array $[1,2,3,3]$ has no majority element: 3 occurs twice, but this isn't enough because it is not true that $2>n/2$.

```Python
def majority(A):
    n = len(A)
    for i in range(n):
        c = 0
        for j in range(n):
            if A[i]==A[j]:
                c = c+1
        if c>n/2:
            return A[i]
    return None
```
